## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/Commuting-Distance-by-Personal-Characteristics/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3], sheet_name='industry')
df
```

The generated data contains the information about commuting distance in Australia in 2016. Column 'Industry (INDP)' includes different industry groups, while 'Average commuting distance (kilometres)' column shows the average commuting distance for these industry groups. The average commuting distance in Australia is 15.98km where Mining industry has the highest average commuting distance (40.28km).   

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(df['Industry (INDP)'])[:-1]
AVG_distances = list(df['Average commuting distance \\n(kilometres)'])[:-1]

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,18))
plt.bar(x_pos, AVG_distances, color='#51BEF8')
plt.xlabel("Industry type", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by Industry", fontsize=28)
plt.ylim((0, 50))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index-0.5, value - 3, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=18, rotation=90)

plt.savefig('img/Commuting distance by industry.pdf', bbox_inches='tight')
plt.show()
```

Across the different industry sectors, people employed in the mining sector reported the longest daily commute with an average commuting distance of 40.3km. By contrast, people employed in both the retail trade and the accommodation and food services sectors report lower than the national average commuting distances (12.94 and 11.41km respectively). Workers employed in the health care and social assistance sector also reported a lower than national average commute of 14.9km.
